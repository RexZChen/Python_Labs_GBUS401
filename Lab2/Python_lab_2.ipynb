# First Step: One - One

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(50, 1)
X
```

```python
def functionY(X, weight, bias):
    return weight * X + bias
```

```python
Y = functionY(X, weight=2, bias=-5)
```

```python
Y
```

```python
noise_seed = np.random.rand(50, 1)
noise = np.sin(noise_seed)
y = Y + noise
```

```python
plt.scatter(X, y)
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
loo = LeaveOneOut()
loo.get_n_splits(X)
```

```python
for train_index, test_index in loo.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print(X_train, X_test, y_train, y_test)
```

```python
plt.scatter(X_train, y_train)
```

```python
plt.scatter(X_test, y_test)
```

```python
from sklearn import linear_model
```

```python
LR = linear_model.LinearRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
predicted_y = LR.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color='black')
plt.plot(X_train, predicted_y, color='blue')
plt.show()
```

```python
y_test
```

```python
LR.predict(X_test)
```

## But how can we tell the performance in more details?

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('Mean squared error: %.2f'
      % mean_squared_error(y_train, predicted_y))
```

```python
print('Coefficient of determination: %.2f'
      % r2_score(y_train, predicted_y))
```

# Second Step: Many - One

```python
feat1 = np.random.rand(200,)
feat2 = np.array([np.random.randint(-100, 100) for _ in range(200)])
feat3 = np.sin(np.random.rand(200,))
random_noise = np.random.rand(200,)
```

```python
def functionY2(feat1, feat2, feat3, weight, bias):
    return weight[0] * feat1 + weight[1] * feat2 + weight[2] * feat3 + bias

y2 = functionY2(feat1, feat2, feat3, weight=[2, -3, 10], bias=-6)
noised_y2 = y2 + random_noise
```

```python
data_set = {
    "feat1": feat1,
    "feat2": feat2,
    "feat3": feat3,
    "y": noised_y2
}
```

```python
from pandas import DataFrame
```

```python
data_set
```

```python
data_set = DataFrame(data_set)
```

```python
data_set.head()
```

```python
data_set.describe()
```

```python
# Our X should be the combination of feat1, feat2 and feat3
X = data_set[["feat1", "feat2", "feat3"]]
y = data_set[["y"]]

print(len(X))
print(len(y))
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
kf.get_n_splits(X)
print(kf)
```

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
LR2 = linear_model.LinearRegression()
```

```python
LR2.fit(X_train, y_train)
```

```python
LR2.coef_
```

```python
LR2.intercept_
```

```python
predicted_y = LR2.predict(X_train)
```

```python
print('Mean squared error: %.2f'
      % mean_squared_error(y_train, predicted_y))
```

```python
print('Coefficient of determination: %.2f'
      % r2_score(y_train, predicted_y))
```

```python
predicted_test = LR2.predict(X_test)
```

```python
y_test - predicted_test
```

## Is this model perfect?

```python
## Overfitting: The model begins to cheat by learning the trainig set distribution
##              usually happens when the model is too easy for the task and fails to figure out the pattern
## Underfitting: The model fails to do a good job; 
##               usually happens when the model is too complex or training resource is limited
```

```python
## Regularization
```

```python
## L2 regularization
LR2_L2 = linear_model.Ridge(alpha=0.5)
LR2_L2.fit(X_train, y_train)
LR2_L2.coef_
```

```python
def sqr_sum(coefs):
    res = 0
    for coef in coefs:
        res += coef ** 2
    
    return sum(res)
```

```python
print(sqr_sum(LR2.coef_))
print(sqr_sum(LR2_L2.coef_))
```

```python
## L1 regularization
LR2_L1 = linear_model.Lasso(alpha=0.5)
LR2_L1.fit(X_train, y_train)
LR2_L1.coef_
```

```python
print(sum(LR2.coef_.squeeze()))
print(sum(LR2_L1.coef_))
```

# More than linear models

```python
N = 800
np.random.seed(3)
x = np.random.uniform(-2,2,N*2).reshape(N,2)
y_ = [1]*N

for i in range(N):
    if x[i][0] * x[i][1] < 0:
        y_[i] = 0
plt.scatter(x[:,0],x[:,1], c=y_)

y = np.array(y_)
y = np.expand_dims(y, axis=1)
print(x.shape)
print(y.shape)
```

```python
from tensorflow import keras
```

```python
NN_Model = keras.models.Sequential([
    keras.layers.Input(shape=(2,)),
    keras.layers.Dense(16, activation='sigmoid'),
    keras.layers.Dense(8, activation='sigmoid'),
    keras.layers.Dense(4, activation='sigmoid'),
    keras.layers.Dense(1, activation='sigmoid')
])
print(NN_Model.summary())
```

```python
opt = keras.optimizers.Adam(learning_rate=1e-3)
```

```python
NN_Model.compile(optimizer=opt, loss='binary_crossentropy',metrics=['accuracy'])
```

```python
NN_Model.fit(x, y,
                epochs=200,
                batch_size=8,
                shuffle=True
                )
```

```python
X = x
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = 0.01
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
print(xx.shape, yy.shape)
Z = NN_Model.predict(np.c_[xx.ravel(), yy.ravel()])
print(Z.shape)
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z,cmap=plt.cm.binary)
plt.scatter(x[:,0],x[:,1], c=y_)
plt.show()
```
